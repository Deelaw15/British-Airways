```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
#create an empty list to collect reviews
reviews = []

#create an empty list to collect Seat type
seat_type = []

#create an empty list to collect stars
stars = []

#create an empty list to collect date
date = []

#create an empty list to collect country
country = []
```

```python
for i in range(1, 40):  
    # Looping through pages
    
    page = requests.get(f"https://www.airlinequality.com/airline-reviews/british-airways/page/{i}/?sortby=post_date%3ADesc&pagesize=100")
    page.raise_for_status()  # Check for HTTP request errors
    soup = BeautifulSoup(page.text, 'html.parser')

    # Extract reviews
    review_items = soup.find_all('div', class_='text_content')
    for item in review_items:
        reviews.append(item.text.strip() if item else "Nan")

    # Extract stars (aligned with reviews)
    star_items = soup.find_all('div', class_="rating-10")[1:]
    for item in range(len(review_items)):
        try:
            if item < len(star_items) and star_items[item].span:
                stars.append(star_items[item].span.text.strip())
            else:
                stars.append("Nan")
        except AttributeError:
            stars.append("Nan")

    # Extract seat types (aligned with reviews)
    seat_type_rows = soup.find_all('td', class_='review-rating-header cabin_flown')
    for item in range(len(review_items)):
        if item < len(seat_type_rows) and seat_type_rows[item].find_next_sibling('td'):
            seat_type.append(seat_type_rows[item].find_next_sibling('td').text.strip())
        else:
            seat_type.append("Nan")

    # Extract dates
    for item in soup.find_all('time'):
        date.append(item.text.strip() if item else "Nan")

    # Extract country information
    for item in soup.find_all('h3'):
        try:
            if item.find('span') and item.find('span').next_sibling:
                country.append(item.find('span').next_sibling.strip("()"))
            else:
                country.append("Nan")
        except:
            print(f"Error extracting country on page {i}")
            country.append("Nan")
```

```python
#Creating a Pandas DataFrame from the list collected during scraping

df = pd.DataFrame({"reviews": reviews, "seat_type": seat_type, "date": date, "country": country, "stars": stars})
```
